```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#Question 5
a=math.sqrt(3)
s1=np.random.uniform(-a,a,1000)
s2=np.random.uniform(-a,a,1000)
plt.scatter(s1, s2)
```

La densité de probabilité jointe du couple (s1,s2) suit une distribution uniforme dans le carré[-sqrt(3),sqrt(3)]*[-sqrt(3),sqrt(3)].
L'indépendance de s1 et s2 peut etre constaté en regardant l'axe x et l'axe y, ils sont séparemment uniformément distribués. 

```python
#Question 6
s1_p=s1.reshape(1,1000)
s2_p=s2.reshape(1,1000)
s=np.concatenate((s1_p, s2_p))
A=np.array([[1,1],[-1,2]])
x=np.dot(A,s)
plt.scatter(x[0], x[1])
```

Etant donné que la matrice A n'est pas une matrice orthogonale. Donc meme s1 et s2 sont indépendants, leur compositions linéaires x1 et x2 ne sont plus indépendants.

```python
W=np.linalg.inv(A)
print(W)
```

```python
Estimation=W.dot(x)
print(np.mean(Estimation-s))
```

En regardant le moyen(très petit) de différence entre les signaux de sources s1,s2 et les estimations calculés en fonction de W, on a bien retrouvé les signaux sources.

On peut retrouver les coefficients de A à partir de la représentation graphique des données, mais juste leurs valeurs absolues, cette limitation est à cause de l'incertitude de correspondance avant et après la projection.

```python
#Question 8 
nb=10000
s1_g=np.random.normal(0,1,nb)
s2_g=np.random.normal(0,1,nb)
plt.scatter(s1_g, s2_g)
```

```python
s1_gp=s1_g.reshape(1,nb)
s2_gp=s2_g.reshape(1,nb)
s=np.concatenate((s1_gp, s2_gp))
x_g=np.dot(A,s)
```

```python
plt.scatter(x_g[0], x_g[1])
```

Dans ce cas c'est pas possible de retrouver les coefficients de A à partir de cette graphique.

```python
#Question 9 
nb=10000
s_gau2=np.random.normal(0,math.sqrt(2),nb)
s_gau3=np.random.normal(0,math.sqrt(3),nb)
s_gau4=np.random.normal(0,math.sqrt(4),nb)

a=math.sqrt(3)
s1=np.random.uniform(-a,a,nb)
s2=np.random.uniform(-a,a,nb)
s3=np.random.uniform(-a,a,nb)
s4=np.random.uniform(-a,a,nb)
```

```python
height_fig = 5
width_fig = 6*height_fig
fig = plt.figure(figsize=(width_fig,height_fig))


ax = plt.subplot(1,6,1)
ax.set_title('s1+s2')
ax.set_ylabel('density')   
ax.set_xlabel('x')
ax.set_xlim(-8, 8)
sns.distplot(s1+s2,bins=25,hist_kws={'log':False},kde=True,ax=ax)

ax = plt.subplot(1,6,2)
ax.set_title('gaussian 2')
ax.set_ylabel('density')  
ax.set_xlabel('x')
ax.set_xlim(-8, 8)
sns.distplot(s_gau2,bins=25,hist_kws={'log':False},kde=True,ax=ax)

ax = plt.subplot(1,6,3)
ax.set_title('s1+s2+s3')
ax.set_ylabel('density')  
ax.set_xlabel('x')
ax.set_xlim(-8, 8)
sns.distplot(s1+s2+s3,bins=25,hist_kws={'log':False},kde=True,ax=ax)

ax = plt.subplot(1,6,4)
ax.set_title('gaussian 3')
ax.set_ylabel('density')  
ax.set_xlabel('x')
ax.set_xlim(-8, 8)
sns.distplot(s_gau3,bins=25,hist_kws={'log':False},kde=True,ax=ax)

ax = plt.subplot(1,6,5)
ax.set_title('s1+s2+s3+s4')
ax.set_ylabel('density')  
ax.set_xlabel('x')
ax.set_xlim(-8, 8)
sns.distplot(s1+s2+s3+s4,bins=25,hist_kws={'log':False},kde=True,ax=ax)

ax = plt.subplot(1,6,6)
ax.set_title('gaussian 4')
ax.set_ylabel('density')  
ax.set_xlabel('x')
ax.set_xlim(-8, 8)
sns.distplot(s_gau4,bins=25,hist_kws={'log':False},kde=True,ax=ax)
```

En comparant ces trois couples de distributions, on a bien constaté que la resemblence entre la somme de plusieurs distributions non gaussiennes et indépendantes et sa distribution gaussiennes correspondantes(meme variance) augmente en fonction de l'augementation du nombre des distributions.
