```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import style
style.use("ggplot")
from sklearn import svm
from sklearn import datasets
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold

#1.1 
A=np.array([[1,5,1],[5,1,1],[1,1,1]])
B= np.array([1,1,-1]).reshape(3,1)
cof=np.linalg.inv(A).dot(B)
print(cof) #corresponds a,b,c
```

```python
X = np.array([[1, 1], [1, 5], [5, 1], [5, 5]])
y = [-1,1,1,1]
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
print(clf.support_vectors_)
print(clf.n_support_)
w = clf.coef_[0]
print("a and b are ",w)
print("c is ", clf.intercept_[0])
a = -w[0] / w[1]
xx = np.linspace(0,6)
yy = a * xx - clf.intercept_[0] / w[1]
plt.plot(xx, yy, 'k-', label="non weighted div")
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.legend()
plt.show()
```

```python
X_3= np.array([[2, 3], [4, 5], [2, 1], [1, 2]])
y_3 = [-1,1,-1,-1]
clf = svm.SVC(kernel='linear')
clf.fit(X_3, y_3)
print(clf.support_vectors_)
print(clf.n_support_)
w = clf.coef_[0]
print("a and b are ",w)
print("c is ", clf.intercept_[0])
a = -w[0] / w[1]
xx = np.linspace(0,6)
yy = a * xx - clf.intercept_[0] / w[1]
plt.plot(xx, yy, 'k-', label="non weighted div")
plt.scatter(X_3[:, 0], X_3[:, 1], c = y_3)
plt.legend()
plt.show()
```

```python
clf = svm.SVC()
iris = datasets.load_iris()
X, y = iris.data, iris.target
#Reduce the dimension of the data to 2D and 3D
X_reduced_2 = PCA(n_components=2).fit_transform(X)
X_reduced_3 = PCA(n_components=3).fit_transform(X)
```

```python
plt.figure(1,figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X_reduced_2[:, 0], X_reduced_2[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('1st eigenvector')
plt.ylabel('2nd eigenvector')

x_min, x_max = X_reduced_2[:, 0].min() - .5, X_reduced_2[:, 0].max() + .5
y_min, y_max = X_reduced_2[:, 1].min() - .5, X_reduced_2[:, 1].max() + .5
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.title("First two PCA directions")

fig = plt.figure(2, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X_reduced_3[:, 0], X_reduced_3[:, 1], X_reduced_3[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
#shuffle the data

X_reduced_2, y = shuffle(X_reduced_2, y, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X_reduced_2, y, test_size=1/3, random_state=42)
```

```python
#2.2 
clf = svm.SVC(gamma = 'auto')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)


cm = confusion_matrix(y_test, y_pred)

def bilanGenerate(cm):
    print(iris.target_names[0])
    print("Precision: ",cm[0][0]/np.sum(cm[:,0]))
    p1 = cm[0][0]/np.sum(cm[:,0])
    print("Recall: ",cm[0][0]/np.sum(cm[0,:]))
    r1 = cm[0][0]/np.sum(cm[0,:])
    print(iris.target_names[1])
    print("Precision: ",cm[1][1]/np.sum(cm[:,1]))
    p2= cm[1][1]/np.sum(cm[:,1])
    print("Recall: ", cm[1][1]/np.sum(cm[1,:]))
    r2 = cm[1][1]/np.sum(cm[1,:])
    print(iris.target_names[2])
    print("Precision: ",cm[2][2]/np.sum(cm[:,2]))
    p3 = cm[2][2]/np.sum(cm[:,2])
    print("Recall: ",cm[2][2]/np.sum(cm[2,:]))
    r3 = cm[2][2]/np.sum(cm[2,:])


def bilanSample(cm):
    p1 = cm[0][0]/np.sum(cm[:,0])
    r1 = cm[0][0]/np.sum(cm[0,:])
    p2= cm[1][1]/np.sum(cm[:,1])
    r2 = cm[1][1]/np.sum(cm[1,:])
    p3 = cm[2][2]/np.sum(cm[:,2])
    r3 = cm[2][2]/np.sum(cm[2,:])
    return np.array([p1, r1, p2, r2, p3, r3]) 

print(bilanGenerate(cm))
```

```python
#2.3 
cv = KFold(n_splits=10, random_state=42, shuffle=True)
bilan = np.zeros((6,))
for train_index, test_index in cv.split(X_reduced_2):
    #print("Train Index: ", train_index, "\n")
    #print("Test Index: ", test_index)

    X_train, X_test, y_train, y_test = X_reduced_2[train_index], X_reduced_2[test_index], y[train_index], y[test_index]
    clf.fit(X_train, y_train) 
    bilan= bilan+ (bilanSample(confusion_matrix(y_test, clf.predict(X_test))))
bilan/=10    
```

```python
print("Setosa: Average Precision and Recall", bilan[0:2])
print("Versicolor: Average Precision and Recall", bilan[2:4])
print("Virginica: Average Precision and Recall", bilan[4:6])
```
