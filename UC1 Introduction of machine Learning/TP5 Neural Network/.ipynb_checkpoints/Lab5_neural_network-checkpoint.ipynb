```python
# Libraries: Standard ones
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import seed
from numpy.random import randn
from numpy.random import uniform
from numpy import linalg as LA
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
# input size
N = 100
# seed the pseudorandom number generator
seed(1255)
# generate random data in which y is a noisy function of x
x = uniform(-5,5,N)
y = x + randn(N) + 3
xplot = np.linspace(-5,5,N);
plt.plot(x,y,"o")
plt.ylabel("y")
plt.xlabel("x")
plt.title("y is a noisy function of x")
plt.show()
```

```python
# Matrix of predictors
x_m = np.copy(x)
# Add column of 1s for intercept coefficient
intercept = np.ones(N)
xx = np.concatenate((intercept[np.newaxis,:],x_m[np.newaxis,:]),axis=0)
xx = np.transpose(xx)
# Initialise w matrix
w = 0.1 * randn(np.shape(xx)[1])
w = w[np.newaxis,:]
w = np.transpose(w)
print(w.shape)
print("xshape=",np.shape(x))
print("yshape=",np.shape(y))
nb_epoch = 200
lr = 0.01
y = y[np.newaxis,:]
y = np.transpose(y)
```

```python
for i in range(nb_epoch):
  w = w + 2*lr*(1/N)*np.sum((y - xx.dot(w))*xx, axis=0, keepdims=True).T
print("The final value of weights are:")
print(w)
```

```python
aline=w[0]+w[1]*xplot
print(aline)
plt.plot(xplot,aline,xplot,x,xplot,y)
plt.legend()
residus = np.dot(xx,w) - y
plt.plot(x,residus,"o")
plt.show()
```

```python
# X = input data
# y = output data
xAll = np.array(([3, 9], [1, 6], [4, 5], [5, 10]), dtype=float)
y = np.array(([18.2], [16], [17.5]), dtype=float)

#
# normalization
xAll = xAll/np.amax(xAll, axis=0)
y = y/20
print("Normalised xAll: ")
print(xAll)
print("Normalised y: ")
print(y)
```

```python
class Neural_Network():
  def __init__(self, inputSize, hiddenSize, outputSize, learning_rate):
    #parameters
    self.inputSize = inputSize
    self.outputSize = outputSize
    self.hiddenSize = hiddenSize
    self.learning_rate = learning_rate
    #add the bias as a neuron always set to 1
    self.inputSize = self.inputSize + 1
    self.hiddenSize = self.hiddenSize + 1
    #weights# weight matrix from input to hidden layer
    self.W1 = np.random.randn(self.inputSize, self.hiddenSize)
    #self.W1 = np.ones((self.inputSize, self.hiddenSize))*0.3
    # weight matrix from hidden to output layer
    self.W2 = np.random.randn(self.hiddenSize, self.outputSize)
    #self.W2 = np.ones((self.hiddenSize, self.outputSize))*0.5

  def feedforward(self, X):
    # forward propagation through our network
    # add the bias to the first input neuron
    bias=np.ones((np.shape(X)[0],1))
    X_bias = np.concatenate((bias, X), axis=1)
    # compute output for hidden layer
    # dot product of X (input) and weight matrix from input to hidden layer
    self.h_pot = np.dot(X_bias, self.W1)
    # activation function
    self.h_out = self.activation_function(self.h_pot)
    # set to 1 the bias neuron on hidden layer
    self.h_out[:,0] = np.ones(np.shape(self.h_out)[0])
    # activation function
    # dot product of hidden layer output (out_h) and weight matrix from hidden to outputlayer
    self.o_pot = np.dot(self.h_out, self.W2)
    # activation function
    self.o_out = self.activation_function(self.o_pot)
    return self.o_out

  def activation_function(self, s):
    # activation function
    return 1/(1+np.exp(-s))

  def backward(self, X, y, output):
    # backward propagate through the network
    # error in output
    self.o_error = y - output
    # applying derivative of activation function to error
    self.o_delta = self.o_error * self.sigmoidPrime(output)
    # h_error: how much our hidden layer weights contributed to output error
    self.h_error = self.o_delta.dot(self.W2.T)
    # applying derivative of activation function to hidden error
    self.h_delta = self.h_error*self.sigmoidPrime(self.h_out)
    bias=np.ones((np.shape(X)[0],1))
    input_layer = np.concatenate((bias, X), axis=1)
    # adjusting first set (input --> hidden) weights
    self.W1 += input_layer.T.dot(self.h_delta) * self.learning_rate
    # adjusting second set (hidden --> output) weights
    self.W2 += self.h_out.T.dot(self.o_delta) * self.learning_rate

  #Question 4
  def sigmoidPrime(self, s):
      return (self.activation_function(s)**2)*np.exp(-s)

  #Question 5
  def learn(self, X, y): 
      out = self.feedforward(X)
      self.backward(X, y ,out)
      return LA.norm(y- self.feedforward(X))
```

```python
#Question 3 with W1 0.3 W2 0.5 
model = Neural_Network(2,3,1,0.01)
print(model.feedforward(xAll)*20)
```

```python
#Question 5
epoch = 200
plot_error = np.zeros(epoch)

x_train = xAll[:3]
y_train = y
x_test = xAll[-1]

# initialization
NN = Neural_Network(2,3,1,0.01)

# trains the NN <<epoch>> times
for i in range(epoch):
  plot_error[i] = NN.learn(x_train, y_train) # Norm ordre 2 

plt.plot(plot_error)
plt.show()
```

```python
#Question 8 
iris = datasets.load_iris()
X = iris.data      # we take all features.
y = iris.target[:, np.newaxis]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42) #Split datasets 
model_iris = Neural_Network(4,10,1, 0.01)
model_iris.learn(X_train, y_train)
```

```python
print("Evaluation sur training datas")
y_traPre = model_iris.feedforward(X_train)
print("The accuracy of training data is")
train_accuracy = np.mean(np.abs(y_train-y_traPre)<0.5) 
print(train_accuracy)
```

```python
print("Evaluation sur testing datas")
y_testPre = model_iris.feedforward(X_test)
print("The accuracy of testing data is")
test_accuracy = np.mean(np.abs(y_test-y_testPre)<0.5) 
print(test_accuracy)
```

After several experiments, we find that the network with only one hidden layer doesn't have a good performance on the iris dataset, so we decide to increase the depth of the neural network, that is what we do in the next part.

```python
# Training data 0.66 testing data 0.33
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) #Split datasets 
model_iris.learn(X_train, y_train)

print("Evaluation sur training datas")
y_traPre = model_iris.feedforward(X_train)
print("The accuracy of training data is")
train_accuracy = np.mean(np.abs(y_train-y_traPre)<0.5) 
print(train_accuracy)
```

```python
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
X_train, X_test, y_train, y_test = train_test_split(X, dummy_y, test_size=0.33, random_state=42)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout


model = Sequential()
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model.fit(X_train, y_train,
          epochs=400,
          batch_size=5)
score = model.evaluate(X_test, y_test, batch_size=5)
```

```python
print("The accurancy of this model(4,16,8,3) is")
print(score[1])
```

The above part shows that a better architecture to cover iris dataset with a test accurancy up to 0.98.

```python
import numpy as np
import matplotlib.pyplot as plt
def twospirals(n_points, noise=.5):
  n = np.sqrt(np.random.rand(n_points,1)) * 780 * (2*np.pi)/360
  d1x = -np.cos(n)*n + np.random.rand(n_points,1) * noise
  d1y = np.sin(n)*n + np.random.rand(n_points,1) * noise
  return (np.vstack((np.hstack((d1x,d1y)),np.hstack((-d1x,-d1y)))),np.hstack((np.zeros(n_points),np.ones(n_points))))
  
X, y = twospirals(1000)
plt.title("training set")
plt.plot(X[y==0,0], X[y==0,1], ".", label="class 1")
plt.plot(X[y==1,0], X[y==1,1], ".", label="class 2")
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# The structure of the network is 2 64 64 64 48 32 16 1 
model = Sequential()
model.add(Dense(64, input_dim=2, activation='tanh'))
model.add(Dense(64, activation='tanh'))
model.add(Dense(64, activation='tanh'))
model.add(Dense(48, activation='tanh'))
model.add(Dense(32, activation='tanh'))
model.add(Dropout(0.5))
model.add(Dense(16, activation='tanh'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train,
          epochs=400,
          batch_size=500)
score = model.evaluate(X_test, y_test, batch_size=5)
```

```python
y_pre = model.predict(X)
y_pre = y_pre>0.5
y_pre = y_pre.reshape(-1,)
```

```python
plt.title("predict set")
plt.plot(X[y_pre==0,0], X[y_pre==0,1], ".", label="class 1")
plt.plot(X[y_pre==1,0], X[y_pre==1,1], ".", label="class 2")
plt.legend()
plt.show()
```
